```python
import folium
import numpy as np
import pandas as pd
import branca
from pyproj import Proj, transform
```

```python
#Import data
df = pd.read_csv('data/data_Rome_4326.csv')

df
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=21,
        tiles='cartodbdark_matter')

# Circle Marker
for i in range(0,len(df)):
   folium.Circle(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       radius=0.5,
       color='orange'
   ).add_to(m)

# Display the map
# m
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=23,
        tiles='cartodbdark_matter')

# Circle Marker
for i in range(0,len(df)):
   folium.Circle(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       radius=0.1,
       weight=0.2,
       color='orange'
   ).add_to(m)

# Display the map
# m
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=23,
        tiles='cartodbdark_matter')

# Circle Marker color-variable
for i in range(0,len(df)):
    if df.iloc[i]['t']=='person':
        color='lightblue'
    elif df.iloc[i]['t']=='bike':
        color='blue'
    elif df.iloc[i]['t']=='car':
        color='red'

    folium.Circle(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       radius=0.2,
       weight=0.2,
       color=color 

    ).add_to(m)

# Display the map
# m
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=21,
        tiles='cartodbdark_matter')

# Popup Marker
for i in range(0,len(df)):
   folium.Marker(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       popup=df.iloc[i]['t']
   ).add_to(m)

# Display the map
# m
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=21,
        tiles='cartodbdark_matter')

# Popup Marker color-variable
for i in range(0,len(df)):
    if df.iloc[i]['t']=='person':
        color='lightblue'
    elif df.iloc[i]['t']=='bike':
        color='blue'
    elif df.iloc[i]['t']=='car':
        color='darkblue'
        
    folium.Marker(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       icon=folium.Icon(color=color)       
    ).add_to(m)
    
# Display the map
# m
```

```python
# Set basemap
m = folium.Map(
        location=[np.mean(df['lat']), np.mean(df['lng'])],
        zoom_start=19,
        max_zoom=21,
        tiles='cartodbdark_matter')

# Popup Marker with fontawesome icons
# check https://fontawesome.com/v4.7.0/ for the full list _ remember to use V4.0 !
for i in range(0,len(df)):
    if df.iloc[i]['t']=='person':
        icon='male',
        color='lightblue'
    elif df.iloc[i]['t']=='bike':
        icon='bicycle'
        color='blue'
    elif df.iloc[i]['t']=='car':
        icon='car',
        color='darkblue'
        
    folium.Marker(
       location=[df.iloc[i]['lat'], df.iloc[i]['lng']],
       icon=folium.Icon(
           prefix='fa',
           icon=icon,
           color=color)       
    ).add_to(m)

# Display the map    
# m
```
